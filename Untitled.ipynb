```python
def category_selection(x):
    sorted_df = x.sort_values(['prediction_confidence'], ascending = False)
    if sorted_df.iloc[0]['prediction_confidence'] > .6:  
        if sorted_df.iloc[0]['predicted_category'] == 'nav':
            #return sorted_df.head(2)
            return sorted_df[sorted_df['prediction_confidence'] > .6]
        else:
            return sorted_df[sorted_df['prediction_confidence'] > .6]
    else:
        if sorted_df.iloc[0]['predicted_category'] == 'nav':
            return sorted_df.head(2) 
        else:
            return sorted_df.head(1)

def compare_y_coord(x):
        if x['quarter_coord'][0] < x['year_coord'][0]:
            return 'left'
        else:
            return 'right'
            
def generalize_quarter_area(cleaned_data_df):
    potential_quarter_data = cleaned_data_df[cleaned_data_df['text'].apply(lambda x: any(y in x.lower() for y in ['period','current','quarter', 'qtd']))]
    potential_year_data = cleaned_data_df[cleaned_data_df['text'].apply(lambda x: any(y in x.lower() for y in ['year', 'ytd', 'itd']))]
    
    potential_groups = set(potential_quarter_data['group_index']) & set(potential_year_data['group_index'])
    potential_quarter_data = potential_quarter_data[potential_quarter_data['group_index'].apply(lambda x: x in potential_groups)].drop_duplicates('combined_block_text')
    potential_year_data = potential_year_data[potential_year_data['group_index'].apply(lambda x: x in potential_groups)].drop_duplicates('combined_block_text')
    #print(' - number of quarter options: ' + str(len(potential_quarter_data)))
    #print(' - number of year options: ' + str(len(potential_year_data)))
    guessed_answer = 'left*'
    
    if len(potential_quarter_data) != 0 and len(potential_year_data) != 0:
        decision_df1 = potential_quarter_data[['group_index','coord_item']].rename({'coord_item':'quarter_coord'}, axis = 1)
        decision_df2 = potential_year_data[['group_index','coord_item']].rename({'coord_item':'year_coord'}, axis = 1)
        merged_decision_df = pd.merge(left = decision_df1, right = decision_df2, on = 'group_index')
        merged_decision_df['decision'] = merged_decision_df.apply(lambda x: compare_y_coord(x), axis = 1)
        #print(' - ' + str(set(merged_decision_df['decision'])))
        if len(set(merged_decision_df['decision'])) == 1:
            guessed_answer = merged_decision_df['decision'][0]
        else:
            guessed_answer = 'left**'

    return guessed_answer


final_df = pd.DataFrame()
cleaner_answers_dict = {}
all_data_dict = {}
'''
file_list = ['2021.09.30.CAB.FTV III.SS Intl IV Guernsey.pdf', '2021.09.30.CAB.FTV Vagaro.HESTA.pdf', '2021.09.30.CAB.Fortress Japan II Yen B.Mesirow Intl.pdf',
            '2021.09.30.CAB.Fortress Japan III Dollar B.CIRS.pdf', '2021.09.30.CAB.HarbourVest IPEP III Partnership.PASERS.pdf', '2021.09.30.CAB.HarbourVest NYSTRS CoInv III.NYSTRS.pdf',
            '2021.09.30.CAB.Insight II.LAFPP.pdf', '2021.09.30.CAB.Lone Star RE III.Bermuda.GOSI.pdf', '2021.09.30.CAB.Madison Inda IV.SSOF II Off.pdf',
            '2021.09.30.CAB.Madison Realty Debt IV.Texas Schools.pdf', '2021.09.30.CAB.Maple H2.NYSTRS.pdf', 
            '2021.09.30.CAB.Maple H3.NYSTRS.pdf','2021.09.30.CAB.New Mountain II.NYCFIRE.pdf', '2021.09.30.CAB.SRE Care.SREP III.pdf']
'''
for file in file_list:
    print(file)
    cleaned_data_df = generate_text_data(pdf_img_dict[file])
    high_confidence_img, best_img, high_confidence_df, best_guess_df, raw_predictions_df = get_predictions(pdf_img_dict[file], cleaned_data_df, lgclf)
    
    high_confidence_df['file'] = [file for x in range(len(high_confidence_df))]
    best_guess_df['file'] = [file for x in range(len(best_guess_df))]
    
    high_confidence_df = high_confidence_df.sort_values(['file','predicted_category', 'prediction_confidence'], ascending=[True, True, False])
    best_guess_df = best_guess_df.sort_values(['file','predicted_category', 'prediction_confidence'], ascending=[True, True, False])

    potential_answers_df = raw_predictions_df.groupby('predicted_category').apply(lambda x: test_method(x)).reset_index(drop = True).sort_values(['predicted_category', 'prediction_confidence'], ascending = [True, False])

    cleaner_answers_df = potential_answers_df[potential_answers_df['prediction_confidence'] > .4]
    if len(cleaner_answers_df) != len(potential_answers_df):
        print(' - dropped some predicted values due to low confidence')
    
    cleaner_answers_df = cleaner_answers_df.sort_values(['predicted_category', 'prediction_confidence'], ascending = [True, False])
    cleaner_answers_dict[file] = cleaner_answers_df
    
    rows_df = cleaned_data_df[cleaned_data_df['group_index'].apply(lambda x: x in set(cleaner_answers_df['group_index']))]
    rows_df['text'] = rows_df['text'].apply(lambda x: '$0' if x == 'SO' else x) # hardcoded rule, might have to add/remove
    rows_df = rows_df[rows_df['text'].apply(lambda x: any(char.isdigit() for char in x) | (x == '-'))] # limits to rows with at least 1 number
    #rows_df = rows_df[rows_df['text'].apply(lambda x: is_date(x, fuzzy = True) == False)] # remove dates
    pulled_values_df = rows_df[['combined_block_text', 'text', 'group_index']]#.groupby('group_index').nth([-1])[['combined_block_text', 'text', 'group_index']] # assuming 1 value
    #pulled_values_df['text'] = pulled_values_df['text'].apply(lambda x: x.replace('S', '$').replace('s', '$').replace('O', '0').replace('o', '0'))

    test_df = pd.merge(
        left = cleaner_answers_df[['preprocessed_text','group_index', 'predicted_category', 'prediction_confidence', 'final_coord_block_par_line_group']],
        right = pulled_values_df,
        on = 'group_index',
        how = 'right'
    )
    test_df = test_df.rename({'text': 'extracted_value'}, axis = 1)
    test_df['file'] = [file for x in range(len(test_df))]
    test_df = test_df.sort_values(['file','predicted_category', 'prediction_confidence'], ascending=[True, True, False])

    # RECATEGORIZING NAVs AFTER NUMBERS HAVE BEEN PULLED; ASSUME FIRST TWO ROWS WITH NUMERIC VALUES ARE THE BEG_NAV/END_NAV
    # TODO: what happens if beg_nav is found and not end_nav or vice versa?
    def nav_categorizer(x, group_set): 
        if x['predicted_category'] != 'nav':
            return True
        else: 
            return x['group_index'] in group_set
    
    top_nav_groups_w_num = set(test_df[test_df['predicted_category'] == 'nav']['group_index'].drop_duplicates()[:2])
    test_df = test_df[test_df.apply(lambda x: nav_categorizer(x, top_nav_groups_w_num), axis =1)] # removing all navs except first two (top 2 confidence)
    
    nav_rows = test_df[test_df['predicted_category'] == 'nav']
    nav_rows['x_coord'] = nav_rows['final_coord_block_par_line_group'].apply(lambda x: x[1])
    nav_rows = nav_rows.sort_values(['x_coord'])
    beg_nav_group_num = nav_rows[['group_index']].drop_duplicates()['group_index'].iloc[0]
    nav_rows['predicted_category'] = nav_rows['group_index'].apply(lambda x: 'beg_nav' if x == beg_nav_group_num else 'end_nav')
    test_df[test_df['predicted_category'] == 'nav'] = nav_rows

    test_df = test_df.drop(['final_coord_block_par_line_group'], axis = 1)
    test_df = test_df.sort_values(['file','predicted_category', 'prediction_confidence'], ascending=[True, True, False])

    test_df = pd.merge(
        left = test_df, 
        right = test_df.groupby('group_index').size().rename('num_cols'), 
        on = 'group_index'
    )
    
    # guessing quarter location (left or right)
    location_guess = generalize_quarter_area(cleaned_data_df)
    print(' - data we want is on the \'' + location_guess + '\' side')
    if 'left' in location_guess:
        narrowed_df = test_df.drop_duplicates('group_index', keep = 'first')
    else:
        narrowed_df = test_df.drop_duplicates('group_index', keep = 'last')

    test_df['guessed_quarter_loc'] = [location_guess for _ in range(len(test_df))]
    all_data_dict[file] = test_df
    
    final_df = pd.concat([final_df, narrowed_df])
```
